## Create a dataframe of the average tempo for each of the top 10 most
# occurring artists.

```python
top_10_artists = df['artist'].value_counts().head(10).index.tolist()
average_tempo_df = pd.DataFrame(columns=['artist', 'average_tempo'])
```

```python
# Calculate the average tempo for each of the top 10 artists
for artist in top_10_artists:
    artist_tempo_mean = df[df['artist'] == artist]['tempo'].mean()
    average_tempo_df = average_tempo_df.append({'Artist': artist, 'Average_Tempo': artist_tempo_mean}, ignore_index=True)
```
